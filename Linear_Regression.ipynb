```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-darkgrid")
# from sklearn.datasets import make_regression
```

```python
# Data generate X,y
```

```python
def generate_dataset(m):
        x = np.random.randn(m)*10
        noise = np.random.randn(m)*12
        print(x.mean(),x.std())
        print(x)
        y =3*x+1+noise
        return x,y
```

```python
x, y = generate_dataset(100)
print(x.shape,y.shape)
```

```python
def plotdata(x,y,color="orange",title = "data"):
    plt.scatter(x,y ,color=color)
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("y",rotation=0, labelpad=20)
    plt.show()
```

```python
plotdata(x,y)
```
